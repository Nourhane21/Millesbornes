On considère le signal analogique $y(t) = cos(2πf0t)$ avec $t$ $∈$ $R$, et $f0 = 3kHz$. On échantillonne $y(t)$ à la fréquence $fe$.
Soit $ye(n)$ le signal ainsi échantillonné.

1) On fixe $fe = 10kHz$. Représenter le signal temporel et le module de sa transformée de Fourier sur deux figures différentes.

```python
import numpy as np
import matplotlib.pyplot as plt

f0 = 3000  # Fréquence du signal analogique en Hz
fe = 10000  # Fréquence d'échantillonnage en Hz
Te = 1 / fe  # Période d'échantillonnage en secondes

# Création du signal échantillonné
t = np.arange(0, 0.01, Te)  # Temps de 0 à 10 ms avec un pas de Te
signal = np.cos(2 * np.pi * f0 * t)  # Signal analogique échantillonné

# Calcul de la transformée de Fourier
N = len(signal)  # Nombre d'échantillons
Y_f = np.fft.fft(signal)  # Transformée de Fourier
frequences = np.fft.fftfreq(N, Te)  # Axe des fréquences

# Tracé du signal temporel
plt.figure(figsize=(12, 6))
plt.subplot(2, 1, 1)
plt.plot(t, signal, 'b-', label='y_e(n)')
plt.title('Signal temporel échantillonné')
plt.xlabel('Temps [s]')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()

# Tracé du module de la transformée de Fourier
plt.subplot(2, 1, 2)
plt.plot(frequences, np.abs(Y_f), 'r-', label='|Y(f)|')
plt.title('Module de la transformée de Fourier')
plt.xlabel('Fréquence [Hz]')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

4) Sous-échantillonner le signal sonore précédent d’un facteur 2 et écoutez-le. Conclure sur les effets du sous-échantionllonnage.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile
from IPython.display import Audio  # Pour écouter le signal dans un environnement Jupyter

# Charger le fichier audio "handel"
fe, data = wavfile.read('/handel.wav')

# Sous-échantillonner d'un facteur 2
facteur_sous_echantillonnage = 2
data_sous_echantillonne = data[::facteur_sous_echantillonnage]
fe_sous_echantillonne = fe // facteur_sous_echantillonnage

# Afficher les informations
print(f"Fréquence d'échantillonnage originale : {fe} Hz")
print(f"Fréquence d'échantillonnage après sous-échantillonnage : {fe_sous_echantillonne} Hz")
print(f"Nombre d'échantillons original : {len(data)}")
print(f"Nombre d'échantillons après sous-échantillonnage : {len(data_sous_echantillonne)}")

# Tracer le signal original et le signal sous-échantillonné
plt.figure(figsize=(12, 8))

# Signal sous-échantillonné
plt.subplot(2, 1, 2)
plt.plot(data_sous_echantillonne, 'r-', label='Signal sous-échantillonné')
plt.title(f'Signal audio sous-échantillonné (fe = {fe_sous_echantillonne} Hz)')
plt.xlabel('Échantillons')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()

print("Écoute du signal sous-échantillonné :")
Audio(data_sous_echantillonne, rate=fe_sous_echantillonne)
```

*Commentaires (effets du sous-échantionllonnage)*:
aprés l'écoute du signal sonore sous-échantillonné, on remarque qu'on a une perte d'information, ainsi qu'une grande distorsion du signal qui peut etre lié à un aliasing

5) Sur-échantillonner le signal sonore précédent d’un facteur 2 et écoutez-le. Conclure sur les effets du sur-échantionllonnage.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile
from scipy.interpolate import interp1d
from IPython.display import Audio

fe, data = wavfile.read('/handel.wav')

# Sur-échantillonner d'un facteur 2
facteur_sur_echantillonnage = 2
n_echantillons_original = len(data) # nombre d'échantillons du signal original
t_original = np.arange(n_echantillons_original)  # Axe temporel original

# Créer un nouvel axe temporel avec plus de points
t_sur_echantillonne = np.linspace(0, n_echantillons_original - 1, n_echantillons_original * facteur_sur_echantillonnage)

# Interpolation pour sur-échantillonner
interpolateur = interp1d(t_original, data, kind='linear', fill_value="extrapolate")
data_sur_echantillonne = interpolateur(t_sur_echantillonne)

# Nouvelle fréquence d'échantillonnage
fe_sur_echantillonne = fe * facteur_sur_echantillonnage

# Afficher les informations
print(f"Fréquence d'échantillonnage originale : {fe} Hz")
print(f"Fréquence d'échantillonnage après sur-échantillonnage : {fe_sur_echantillonne} Hz")

# Tracer le signal sur-échantillonné
plt.figure(figsize=(12, 8))

# Signal sur-échantillonné
plt.subplot(2, 1, 2)
plt.plot(t_sur_echantillonne, data_sur_echantillonne, 'r-', label='Signal sur-échantillonné')
plt.title(f'Signal audio sur-échantillonné (fe = {fe_sur_echantillonne} Hz)')
plt.xlabel('Échantillons')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()

print("Écoute du signal sur-échantillonné :")
Audio(data_sur_echantillonne, rate=fe_sur_echantillonne)
```

*Commentaires (effets du sur-échantionllonnage)*: on remarque une bonne représentation du signal et on arrive beaucoup plus à avoir plus de détails qui étaient masqués auparavnt, moins d'effets de distorsion

Soit le signal $x(n) = (1/2)^n$  $u(n)$, avec $u(n)$ l’échelon unité.

1) Calculer $X(f)$ la transformée de Fourier de $x(n)$, avec $n = −128, ..., −1, 0, 1, 2, ..., 128$. Représenter le module et la phase de $X(f)$. On rappelle que $f$ $∈$ $[$−1/2$,  $1/2$]$



```python
import numpy as np
import matplotlib.pyplot as plt

# Définir le signal x(n)
n = np.arange(0, 129)  # n = 0, 1, 2, ..., 128
x = (1/2)**n  # x(n) = (1/2)^n

# Calculer la transformée de Fourier discrète (TFD)
N = len(n)
f = np.fft.fftfreq(N, d=1)  # Fréquences normalisées (f ∈ [-0.5, 0.5])
X_f = np.fft.fft(x)  # TFD de x(n)

# Module et phase de X(f)
module_X_f = np.abs(X_f)  # Module de X(f)
phase_X_f = np.angle(X_f)  # Phase de X(f)

# Tracé du module de X(f)
plt.figure(figsize=(12, 6))
plt.subplot(2, 1, 1)
plt.plot(f, module_X_f, 'b-', label='|X(f)|')
plt.title('Module de la transformée de Fourier')
plt.xlabel('Fréquence normalisée (f)')
plt.ylabel('|X(f)|')
plt.grid(True)
plt.legend()

# Tracé de la phase de X(f)
plt.subplot(2, 1, 2)
plt.plot(f, phase_X_f, 'r-', label='Phase de X(f)')
plt.title('Phase de la transformée de Fourier')
plt.xlabel('Fréquence normalisée (f)')
plt.ylabel('Phase (radians)')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
